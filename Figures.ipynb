```python
fig,ax = plt.subplots(figsize=(10,10))

ax.plot([1,9,3,1],[1,2,8,1])
ax.plot([1,4],[1,4])
ax.plot([4,9],[4,2])
ax.plot([4,3],[4,8])

ax.text(1,.6,'A',fontsize=18)
ax.text(9.2,1.8,'B',fontsize=18)
ax.text(2.5,8,'C',fontsize=18)
ax.text(4.2,4,'X',fontsize=18)

ax.text(4,2,'area C',fontsize=18)
ax.text(4.3,5,'area A',fontsize=18)
ax.text(2.2,4,'area B',fontsize=18)

ax.set_axis_off()

fig.savefig('data/bc1.png')
```
